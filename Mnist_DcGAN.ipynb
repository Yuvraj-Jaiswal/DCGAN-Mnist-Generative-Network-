```python
from tensorflow.keras.layers import *
from tensorflow.keras.datasets import mnist
from tensorflow.keras import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np

BatchSize = 64

(train_images , _ ) , (_,_) = mnist.load_data()
train_images = train_images.reshape(60000,28,28,1)

train_images = (train_images.astype('float32')-127.5)/127.5
noice_size = 100

dataSet = tf.data.Dataset.from_tensor_slices(train_images).shuffle(60000).batch(128 , drop_remainder=True)

Generator = Sequential()
Generator.add(Dense(7*7*128 , input_shape=(noice_size,)))  ## shape [7,7,128]
Generator.add(Reshape(target_shape=(7,7,128)))

Generator.add(Conv2DTranspose(128 , (5,5) , strides=(2,2) , padding='same' , activation=LeakyReLU()))
## shape [14,14,128]

Generator.add(Conv2DTranspose(64 , (5,5) , strides=(2,2) , padding='same' , activation=LeakyReLU()))
## shape [28,28,64]

Generator.add(Conv2DTranspose(1 , (5,5) , padding='same' , activation='tanh'))
## shape [28,28,1]


Detector = Sequential()
Detector.add(Conv2D(64, (5,5), strides=(2, 2), padding='same'))
Detector.add(LeakyReLU())

Detector.add(Conv2D(128, (5,5), strides=(2, 2), padding='same'))
Detector.add(LeakyReLU())

Detector.add(Flatten())
Detector.add(Dense(1,activation="sigmoid"))
Detector.compile(loss='binary_crossentropy' ,optimizer=Adam(lr=0.0002, beta_1=0.5), metrics=['accuracy'])
Detector.trainable = False


GAN_Model = Sequential([Generator,Detector])
GAN_Model.compile(optimizer=Adam(lr=0.0002, beta_1=0.5) , loss='binary_crossentropy' , metrics=['accuracy'])

Epoches = 200


for epoch in range(Epoches):

    print(f"At Epoch : {epoch+1} / {Epoches}")

    Batch_Count = 0

    Noise_Sample = tf.random.normal((12, noice_size))
    Generated_Image_Sample = Generator.predict(Noise_Sample)

    plt.figure(figsize=(10,10))

    for i in range(12):
      plt.subplot(4,4,i+1)
      plt.imshow(Generated_Image_Sample[i].reshape(28,28),cmap="binary")
      plt.axis('off')
    plt.show()

    for Image_Batch in dataSet:

        Half_Batch = BatchSize//2

        if Batch_Count%100==0:
            print(f"At Batch {Batch_Count} >>>>>")
            print(f"Detector - {det_info} , Generator - {gan_info}")


        Noise = tf.random.normal((Half_Batch,noice_size))
        Generated_Image = Generator.predict(Noise)

        Combined_Images = tf.concat([Generated_Image,tf.dtypes.cast(Image_Batch[0:Half_Batch] , tf.float32)], axis=0)
        
        Combined_Labels = tf.constant([[0]]*Half_Batch + [[1]]*Half_Batch)

        Detector.trainable = True
        
        det_info = Detector.train_on_batch(Combined_Images , Combined_Labels)

        Detector.trainable = False
        
        Noise2 = tf.random.normal((BatchSize , noice_size))

        Labels = tf.constant([[1]]*BatchSize)

        gan_info = GAN_Model.train_on_batch(Noise2,Labels)


        Batch_Count += 1
```

```python
Generator.save("Generator-MNIST.h5")
```
